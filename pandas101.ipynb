## วนลูป record ในตาราง (iterrows)

```python
unit.iterrows()
```

```python
for row in unit.iterrows():
    print(row)
```

```python
for row in unit.head().iterrows():
    if 'project' not in row[1]['project_id']:
    print(row)
```

```python
for row in unit.iterrows():
    if 'สตูดิโอ' in row[1]['url_unittype']:
        print(row[1]['url_unittype'])
```

```python
for row in unit.head(2).iterrows():
    print(row)
    print('จำนวนสมาชิก',len(row))
    print('จำนวนสมาชิก 1',row[0])
    print('จำนวนสมาชิก 2',row[1])
```
